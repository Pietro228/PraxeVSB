- We need to somehow mark the outliers in the data according to the *SalePrice* and *GrLivArea*

    - Lower bound: Q1 - 1.5*IQR

```python
#prep
import pandas as pd # dataframes
import numpy as np # matrices and linear algebra
import matplotlib.pyplot as plt # plotting
import seaborn as sns # another matplotlib interface - styled and easier to use

df_full = pd.read_csv("https://raw.githubusercontent.com/rasvob/VSB-FEI-Fundamentals-of-Machine-Learning-Exercises/master/datasets/zsu_cv1_data.csv")

print("Hello")
```

```python
#HW1
df = df_full
#IQR = Q3 - Q1, Interquartile range, range with higher tolarance for outliers
SP_Q1 = df["SalePrice"].quantile(0.25)
SP_Q3 = df["SalePrice"].quantile(0.75)
SP_LB = SP_Q1 - 1.5*(SP_Q3 - SP_Q1)
SP_UB = SP_Q3 + 1.5*(SP_Q3 - SP_Q1)

GLA_Q1 = df["GrLivArea"].quantile(0.25)
GLA_Q3 = df["GrLivArea"].quantile(0.75)
GLA_LB = GLA_Q1 - 1.5*(GLA_Q3 - GLA_Q1)
GLA_UP = GLA_Q3 + 1.5*(GLA_Q3 - GLA_Q1)

df["Outlier"] = False
df.loc[(df.SalePrice < SP_LB) | (df.SalePrice > SP_UB) | (df.GrLivArea < GLA_LB) | (df.GrLivArea > GLA_UP), "Outlier"] = True

plt.figure(figsize=(12,18))
sns.scatterplot(data = df, x='SalePrice', y='GrLivArea', hue='Outlier')
```

```python
#HW2
#cutting into bins for better vis
df['SalePriceBins'] = pd.cut(df['SalePrice'], bins=8)


#visualization of SalePrice and OverallQual
#plt.figure(figsize=(6,4)) #width, height
#subplots ret a tuple, below is an actuall valid decl and init of a tuple in Python
fig, axes = plt.subplots(1,2, figsize=(20,10)) #(rowCnt, colCnt, entire figsize)
#barplot
sns.barplot(data=df, x='SalePriceBins', y='OverallQual', ax=axes[0])
#boxplot
sns.boxplot(data=df, x='SalePriceBins', y='OverallQual', ax=axes[1])
#plt.xticks(rotation=45) #spins only for last, even when duplicated and shifted
for ax in axes: #rotates for all axis
  ax.tick_params(axis='x', rotation=45) #tick_params, cause axes do not have xtick
```
